# Check Backend Jobs Queue

```python
import os
from qiskit_ionq import IonQProvider
```

```python
from dotenv import load_dotenv

load_dotenv()
provider = IonQProvider(os.getenv('IONQ_API_KEY'))
```

```python
# 1. Get the specific backend first
backend = provider.get_backend("ionq_simulator")

# 2. Retrieve jobs from THAT backend
jobs = backend.jobs(limit=5)

if not jobs:
    print("No recent jobs found for this backend.")
else:
    print(f"--- Recent Jobs for {backend.name()} ---")
    for job in jobs:
        print(f"ID: {job.job_id()} | Status: {job.status()}")
```
